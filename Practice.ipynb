```python
# Importing the dependencies
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import re
import requests
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Quotes to Scrape site
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title
title = html_soup.find('h2').text
title
```

```python
# Scrape the top ten tags
tag_box = html_soup.find('div', class_='tags-box')
# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
# Browsing the page
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Capturing and printing the quotes from the first 5 pages
for x in range(1, 6):
    html = browser.html
    quote_soup = soup(html, 'html.parser')
    quotes = quote_soup.find_all('span', class_='text')
    for quote in quotes:
        print('page:', x, '----------')
        print(quote.text)
    browser.links.find_by_partial_text('Next').click()
```

```python
# Capturing and printing the quotes from the first 5 pages
for x in range(1, 6):
    html = browser.html
    quote_soup = soup(html, 'html.parser')
    quotes = quote_soup.find_all('div', class_='quote')
    for quote in quotes:
        print('page:', x, '----------')
        print(quote.text)
    if x < 5:
        browser.links.find_by_partial_text('Next').click()
```

```python
booktitles = books_soup.find_all('div', class_='image_container')
print(booktitles)
```

```python
for booktitle in booktitles:
    title = booktitle.img['alt']
    print(title)
```

```python
baseurl = 'https://books.toscrape.com/'
for booktitle in booktitles:
    image_url = booktitle.img['src']
    title = booktitle.img['alt']
    print(f'{baseurl}{image_url} {title}')
```

```python
# Capture all info regarding books
books_info = books_soup.find_all('div', class_='row')
books_info2 = requests.get("https://books.toscrape.com/")
books_info2.text
```

```python
# Defining the regular expression to capture pricess in £
# Adapting code found here: https://blog.hartleybrody.com/web-scraping-cheat-sheet/
prices = re.findall(r'£[0-9,.]+', books_info2.text)
for price in prices:
    print(price)
```
